# Train/Testデータの中身を確認する

```python
import pandas as pd
```

```python
# Data load
titanic_train = pd.read_csv('./data/train.csv')
titanic_test = pd.read_csv('./data/test.csv')
```

# データの中身
* Pclass, Sex, Age, SibSp, Parch, Fare, Embarkedは数値 or カテゴリデータで扱いやすそう
* Nameはテキストデータでそのままでは使いづらい
* Ticketもバリエーションが多くで加工が必要
* Cabinは欠損が多いのとバリエーションが多く加工が必要

# 欠損値の確認
## 学習データ
* Age: 177 (19.9%)
* Cabin: 687 (77.1%)
* Embarked: 2 (0.2%)
  * 学習データ中にのみ欠損あり

## 評価データ
* Age: 86 (20.6%)
* Fare: 1 (0.2%)
  * 評価データ中にのみ欠損あり
* Cabin: 327(78.3%)

```python
nan_df = pd.concat({'count': titanic_train.isnull().sum(), 'rate': 100*titanic_train.isnull().mean()}, axis=1)
nan_df.query('count > 0')
```

```python
nan_df = pd.concat({'count': titanic_test.isnull().sum(), 'rate': 100*titanic_test.isnull().mean()}, axis=1)
nan_df.query('count > 0')
```
